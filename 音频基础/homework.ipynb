# 3.波表合成

## 乐音波形上下颠倒

```python
y1, sr1 = librosa.load('music.wav')
for i in range(y1.size):
    y1[i] = -y1[i]
librosa.display.waveplot(y1, sr=sr1)
ipd.Audio(y1, rate=sr1)
```

## 乐音波形左右颠倒

```python
y1, sr1 = librosa.load('music.wav')
y1_ = np.zeros(y1.size)
for i in range(y1.size):
    y1_[i] = y1[(-1*i)%y1.size]
librosa.display.waveplot(y1_, sr=sr1)
ipd.Audio(y1_, rate=sr1)
```

## 人声波形左右颠倒

```python
y2, sr2 = librosa.load('speak.wav')
y2_ = np.zeros(y2.size)
for i in range(y2.size):
    y2_[i] = y2[(-1*i)%y2.size]
librosa.display.waveplot(y2_, sr=sr2)
ipd.Audio(y2_, rate=sr2)
```

## 给乐音设定不同的阈值

```python
y1, sr1 = librosa.load('music.wav')
ymin, ymax = y1.min(), y1.max()
print(ymin,ymax)
y1_ = (y1-ymin)/(ymax-ymin)
librosa.display.waveplot(y1_, sr=sr1)
ipd.Audio(y1_, rate=sr1)

for i in range(y1_.size):
    if y1_[i]>0.7:
        y1_[i]=0.7
    if y1_[i]<-0.7:
        y1_[i]=-0.7
librosa.display.waveplot(y1_, sr=sr1)
ipd.Audio(y1_, rate=sr1)

for i in range(y1_.size):
    if y1_[i]>0.5:
        y1_[i]=0.5
    if y1_[i]<-0.5:
        y1_[i]=-0.5
librosa.display.waveplot(y1_, sr=sr1)
ipd.Audio(y1_, rate=sr1)

for i in range(y1_.size):
    if y1_[i]>0.3:
        y1_[i]=0.3
    if y1_[i]<-0.3:
        y1_[i]=-0.3
librosa.display.waveplot(y1_, sr=sr1)
ipd.Audio(y1_, rate=sr1)

for i in range(y1_.size):
    if y1_[i]>0.1:
        y1_[i]=0.1
    if y1_[i]<-0.1:
        y1_[i]=-0.1
librosa.display.waveplot(y1_, sr=sr1)
ipd.Audio(y1_, rate=sr1)
```

## 给人声设定不同的阈值

```python
y2, sr2 = librosa.load('speak.wav')
ymin, ymax = y2.min(), y2.max()
print(ymin,ymax)
y2_ = (y2-ymin)/(ymax-ymin)
librosa.display.waveplot(y2_, sr=sr2)
ipd.Audio(y2_, rate=sr2)

for i in range(y2_.size):
    if y2_[i]>0.7:
        y2_[i]=0.7
    if y2_[i]<-0.7:
        y2_[i]=-0.7
librosa.display.waveplot(y2_, sr=sr2)
ipd.Audio(y2_, rate=sr2)

for i in range(y2_.size):
    if y2_[i]>0.5:
        y2_[i]=0.5
    if y2_[i]<-0.5:
        y2_[i]=-0.5
librosa.display.waveplot(y2_, sr=sr2)
ipd.Audio(y2_, rate=sr2)

for i in range(y2_.size):
    if y2_[i]>0.3:
        y2_[i]=0.3
    if y2_[i]<-0.3:
        y2_[i]=-0.3
librosa.display.waveplot(y2_, sr=sr2)
ipd.Audio(y2_, rate=sr2)

for i in range(y2_.size):
    if y2_[i]>0.1:
        y2_[i]=0.1
    if y2_[i]<-0.1:
        y2_[i]=-0.1
librosa.display.waveplot(y2_, sr=sr2)
ipd.Audio(y2_, rate=sr2)
```
