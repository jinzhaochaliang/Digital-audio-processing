```python
import librosa
x,sr=librosa.load('whatMovies.wav',sr=8000)
N=x.size
y=(x[0:N-1]+x[1:N])/2
x=x[0:N-1]
import IPython.display as ipd
print('original audio')
ipd.display(ipd.Audio(x,rate=sr))
print('after two point average')
ipd.display(ipd.Audio(y,rate=sr))
```

```python
import numpy as np
T=2
sr=8000
N=int(T*sr)
noise=np.random.rand(N+1)-0.5
x=noise[0:N]
y=(noise[0:N]+noise[1:N+1])/2
import IPython.display as ipd
print('original audio')
ipd.display(ipd.Audio(x,rate=sr))
print('after two point average')
ipd.display(ipd.Audio(y,rate=sr))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
xz=np.abs(np.fft.rfft(x))
yz=np.abs(np.fft.rfft(y))
f=np.arange(xz.size)/xz.size*sr/2
fig,ax=plt.subplots(3)
ax[0].set_title('X[Z]')
ax[0].plot(f,xz)
ax[1].set_title('Y[Z]')
ax[1].plot(f,yz)
ax[2].set_title('H[Z]')
ax[2].plot(f,yz/xz)
plt.tight_layout()
plt.show()
```
